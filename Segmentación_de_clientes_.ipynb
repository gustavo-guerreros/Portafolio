```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Análisis exploratorio de datos
# Obtener información sobre las características de los clientes y su comportamiento
print("Información sobre las características de los clientes y su comportamiento:")
train_df.describe()

```

```python
# Explorar las relaciones entre las variables
print("\nRelaciones entre las variables:")
correlation_matrix = train_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()


```

```python
# Segmentación de clientes
# Preparar los datos para la segmentación
X = train_df[['Age', 'Work_Experience', 'Family_Size']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Realizar la segmentación utilizando K-Means
n_clusters = 3  # Número de clusters que deseamos obtener
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
train_df['Cluster'] = kmeans.fit_predict(X_scaled)

```

```python
# Visualizar la segmentación
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=train_df['Cluster'], cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Work_Experience')
plt.title('Segmentación de Clientes')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersión para visualizar la segmentación
plt.figure(figsize=(8, 6))

# Plot puntos de cada cluster
for cluster_num in range(n_clusters):
    cluster_points = X_scaled[train_df['Cluster'] == cluster_num]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_num + 1}')

# Plot centroide de cada cluster
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=100, c='red', marker='X', label='Centroides')

# Añadir etiquetas y leyendas
plt.xlabel('Age (Scaled)')
plt.ylabel('Work_Experience (Scaled)')
plt.title('Segmentación de Clientes utilizando K-Means')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Evaluación y selección de segmentos
# Calcular el índice de silueta para evaluar la calidad de la segmentación
silhouette_avg = silhouette_score(X_scaled, train_df['Cluster'])
print(f"\nÍndice de silueta: {silhouette_avg}")

# Seleccionar los segmentos más relevantes y significativos para el análisis
# Por ejemplo, puedes analizar las características de cada segmento y compararlas para tomar una decisión.
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Preparar los datos para la segmentación
X_train = train_df[['Age', 'Work_Experience', 'Family_Size']]
X_test = test_df[['Age', 'Work_Experience', 'Family_Size']]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Realizar la segmentación utilizando K-Means
n_clusters = 3  # Número de clusters que deseamos obtener
n_init = 10  # Número de veces que se ejecutará K-Means con diferentes centroides iniciales
kmeans = KMeans(n_clusters=n_clusters, n_init=n_init, random_state=42)
train_df['Cluster'] = kmeans.fit_predict(X_train_scaled)
test_df['Cluster'] = kmeans.predict(X_test_scaled)
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Obtener las características utilizadas para la segmentación
X = train_df[['Age', 'Work_Experience', 'Family_Size']]

# Calcular el puntaje de silueta y el índice de Calinski-Harabasz para evaluar la calidad de la segmentación
labels = train_df['Cluster']
silhouette_score_value = silhouette_score(X, labels)
calinski_harabasz_score_value = calinski_harabasz_score(X, labels)

print("Puntaje de silueta:", silhouette_score_value)
print("Índice de Calinski-Harabasz:", calinski_harabasz_score_value)
```

```python
import matplotlib.pyplot as plt

# Graficar los clusters en un gráfico de dispersión
plt.scatter(X['Age'], X['Work_Experience'], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Work_Experience')
plt.title('Segmentación de clientes con K-Means')
plt.show()

# También puedes usar gráficos de histogramas para visualizar la distribución de las características en cada cluster
plt.hist(X[labels == 0]['Family_Size'], alpha=0.5, label='Cluster 0', color='red')
plt.hist(X[labels == 1]['Family_Size'], alpha=0.5, label='Cluster 1', color='blue')
plt.hist(X[labels == 2]['Family_Size'], alpha=0.5, label='Cluster 2', color='green')
plt.xlabel('Family_Size')
plt.ylabel('Frecuencia')
plt.title('Distribución de Family_Size en cada cluster')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersión para los datos de entrenamiento
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=train_df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Work_Experience')
plt.title('Segmentación de clientes - Gráfico de dispersión')
plt.colorbar(label='Cluster')
plt.show()
```
